```
#include <stdio.h>
#include <stdlib.h>
#include <time.h>


void FindInArray(int *p, int *min, int *max, float *aver, int N)
{
    *min=p[0];
    *max=p[0];
    int sum=0;

    for (int i=0; i<N; i++)
    {
        sum=sum+p[i];
    }

    for (int i=0; i<N; i++)
    {
        if (p[i]>*max)
        {
            *max=p[i];
        }
        if (p[i]<*min)
        {
            *min=p[i];
        }
    }
    *aver=sum/(float)N;

}

void Printing (int p[], int N)
{
    for(int i=0; i<N; i++)
    {
        printf("arr[%d]: %d\n", i, p[i]);
    }
}

int main()
{
    int arr[20];
    int i;
    srand(time(NULL));

    for (i=0; i<20; i++)
    {
        arr[i]=rand()%1000;
    }

    Printing (arr, 20);
   int *minimum;
   int *maximum;
   float *average;
   FindInArray(arr, minimum, maximum, average,20);
   printf("Smallest number: %d\n", *minimum);
   printf("Biggest number: %d\n", *maximum);
   printf("The average:  %f\n", *average);


    return 0;
}
```
