```
#include <stdio.h>
#include <stdlib.h>
#include <time.h>



void DisplayArray (int N, int A[][100])
{
    /*Counter variables for the loop*/
    int i,j;
    for (i=0; i<N; i++)
    {
        for (j=0; j<N; j++)
        {
            printf("%d\t", A[i][j]); /*Displaying array elements*/
        }
        printf("\n");
    }
}

void  FindAverageOfRow (int N, int A[][100], float B[])
{
    int i, j;
    /* calculate the average of rows of the array using for loop*/
    for (i=0; i<N; i++)
    {
        for (j=0; j<N; j++)
        {
            B[i]=B[i]+A[i][j]; 
        }
        B[i]=B[i]/N;
    }
}
int main()
{

    int matrix [100][100];
    float RowAverOfArray[100]={0};
    int size;
    srand(time(NULL));  //initialize the random seed


    printf("Input the size of the square matrix: ");
    scanf("%d", &size);

    int i,j;
    for (i=0; i<size; i++)
    { for (j=0; j<size; j++)
    {
        matrix[i][j]=rand()%900+100; 
    }
    }

    DisplayArray (size, matrix);
    FindAverageOfRow(size, matrix, RowAverOfArray);
    for (i=0; i<size; i++)
    {
        printf("%4.2f", RowAverOfArray);
    }

    return 0;
}

```
